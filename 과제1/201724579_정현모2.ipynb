```python
INPUT_STR = "The Turing test, originally called the imitation game by Alan Turing in 1950, is a test of a machine's ability to exhibit intelligent behaviour equivalent to, or indistinguishable from, that of a human."
```

```python
def strToOneHotFunc(INPUT_STR):
    new_string = ""
    for c in INPUT_STR:
        if c != '.' and c != ',':
            new_string += c

    vector_list = new_string.split()
    
    vector_map = {}
    index = 0

    for vector in vector_list:
        if vector not in vector_map:
            vector_map[vector] = index
            index += 1

    print("각 단어 encoding")
    print(vector_map)

    for key, value in vector_map.items():
        one_hot_encoding = [0 for _ in range(len(vector_list))]
        one_hot_encoding[value] = 1
        vector_map[key] = one_hot_encoding

    result = []
    for vector in vector_list:
        result.append(vector_map[vector])
    
    return result
```

```python
result = strToOneHotFunc(INPUT_STR)
```
