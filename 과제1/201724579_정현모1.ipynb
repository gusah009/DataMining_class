```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras import datasets, layers, models
```

```python
train_df = pd.read_csv("../Ex1_dataset/emnist-letters-train.csv",header=None)
test_df = pd.read_csv("../Ex1_dataset/emnist-letters-test.csv",header=None)
train_df
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
train_labels = (train_df[0] - 1).to_numpy()
test_labels = (test_df[0] - 1).to_numpy()
```

```python
train_images = train_df.iloc[:, 1:]
test_images = test_df.iloc[:, 1:]
train_images
```

```python
train_numpy_images = train_images.to_numpy().reshape((train_images.shape[0], 28 ,28, 1))
test_numpy_images = test_images.to_numpy().reshape((test_images.shape[0], 28 ,28, 1))
train_numpy_images.shape
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (7, 7), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((5, 5), padding="same"))
model.add(layers.Conv2D(64, (2, 2), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (2, 2), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(26, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_numpy_images, train_labels, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(test_numpy_images,  test_labels, verbose=2)
```

```python
CHOICE_COUNT = 10
```

```python
choice_random_index = np.random.choice(test_images.shape[0], CHOICE_COUNT, replace=False)
choice_test_images = test_numpy_images[choice_random_index]
choice_test_images.shape
```

```python
predict_test_labels = model.predict(choice_test_images).argmax(axis=1)
predict_test_labels
```

```python
choice_test_labels = test_labels[choice_random_index]
choice_test_labels
```

```python
accuracy = sum(predict_test_labels == choice_test_labels) / CHOICE_COUNT * 100
print(f"accuracy: {accuracy}%")
```
